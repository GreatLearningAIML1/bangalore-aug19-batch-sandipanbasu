## Question 1

### Import dataset
- Import iris dataset

```python
# Import pandas 
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Read in white wine data 
iris = load_iris()
iris
```

## Question 2

### Get features and label from the dataset in separate variable
- you can get the features using .data method
- you can get the features using .target method

```python
feature_set = iris.data
target_set = iris.target
```

## Question 3

### Create train and test data
- use train_test_split to get train and test set
- set a random_state
- test_size: 0.25

```python
from sklearn.model_selection import train_test_split
# Split the data up in train and test sets
X_train, X_test, y_train, y_test = train_test_split(feature_set, target_set, test_size=0.25, random_state=600)
```

## Question 4

### One-hot encode the labels
- convert class vectors (integers) to binary class matrix
- convert labels
- number of classes: 3
- we are doing this to use categorical_crossentropy as loss

```python
y_train_binary = keras.utils.to_categorical(y_train, num_classes=None)
y_train_binary

y_test_binary = keras.utils.to_categorical(y_test, num_classes=None)
y_test_binary
```

## Question 5

### Initialize a sequential model
- Define a sequential model

```python
# Initialize the constructor
model = tf.keras.Sequential()
```

## Question 6

### Add a layer
- Use Dense Layer  with input shape of 4 (according to the feature set) and number of outputs set to 3
- Apply Softmax on Dense Layer outputs

```python
# Add an input layer 
model.add(tf.keras.layers.Dense(10, input_shape=(4,)))

# Add an output layer 
model.add(tf.keras.layers.Dense(3, activation='softmax'))
```

## Question 7

### Compile the model
- Use SGD as Optimizer
- Use categorical_crossentropy as loss function
- Use accuracy as metrics

```python
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
print(model)
print(model.optimizer)
print(model.loss)
print(model.metrics_names)
```

## Question 8

### Summarize the model
- Check model layers
- Understand number of trainable parameters

```python
model.summary()
```

## Question 9

### Fit the model
- Give train data as training features and labels
- Epochs: 100
- Give validation data as testing features and labels

```python
history = model.fit(X_train, y_train_binary,epochs=100, batch_size=1, verbose=1)
```

```python
score = model.evaluate(X_test, y_test_binary,verbose=1)

print(score)
```

## Question 4

### Get features and label from the dataset in separate variable
- Take "open", "close", "low", "high" columns as features
- Take "volume" column as label
- Normalize label column by dividing it with 1000000

```python
feature_set = prices.drop('volume',axis=1)
target_set = prices[['volume']].copy()
target_set.columns = ['volume']
target_set = target_set / 1000000

target_set.head()
```

## Question 5

```python
feature_set.info()
```

## Question 6

### Normalize data
- Normalize features
- Use tf.math.l2_normalize to normalize features
- You can read more about it here https://www.tensorflow.org/api_docs/python/tf/math/l2_normalize

```python
feature_l2nor= tf.math.l2_normalize(feature_set_arr.astype('float32'))
feature_l2nor
```

## Question 7

### Define weight and bias
- Initialize weight and bias with tf.zeros
- tf.zeros is an initializer that generates tensors initialized to 0
- Specify the value for shape

```python
weight = tf.Variable(tf.zeros([4,1]))
bias = tf.Variable(tf.zeros([1]))

weight
```

## Question 8

### Get prediction
- Define a function to get prediction
- Approach: prediction = (X * W) + b; here is X is features

```python
def prediction(x_train,w, b):
    p = tf.matmul(x_train,w) + b
    return p
    
    
pred = prediction(feature_l2nor,weight, bias)
#print(pred)
```

### Calculate loss
- Calculate loss using predictions
- Define a function to calculate loss
- We are calculating mean squared error

```python
def loss(error):
    # print('inside error func ')
    loss = tf.reduce_mean(tf.square(error))
    # print(loss)
    return loss;
```

## Question 9

### Define a function to train the model
1.   Record all the mathematical steps to calculate Loss
2.   Calculate Gradients of Loss w.r.t weights and bias
3.   Update Weights and Bias based on gradients and learning rate to minimize loss

```python
#training info
train_steps = 100
learning_rate = 0.01

for i in range(train_steps):
    #watch the gradient flow 
    with tf.GradientTape() as tape:
        #forward pass 
        yhat = prediction(feature_l2nor,weight,bias)            
        #calcuate the loss (difference squared error)
        error = tf.math.subtract(yhat,tf.convert_to_tensor(target_set_arr, dtype=tf.float32))
        l = loss(error)

    #dw dbias 
    dw, db = tape.gradient(l, [weight, bias])

    #update the weights and biases 
    weight.assign_sub(dw * learning_rate)
    bias.assign_sub(db* learning_rate)
    
    #print the loss every 10 iterations 
    if i % 10 == 0:
        print("Loss at step {:03d}: {:.3f}".format(i, l))
      
print(f'W : {weight.numpy()} , b  = {bias.numpy()} ')
```
